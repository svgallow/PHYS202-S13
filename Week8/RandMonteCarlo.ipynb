Example 1: Ananlytic Inverse CDF Transform
-

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0) -cos(x))/2.

plot(x, f(x), label='True distribution', lw = 3)
plot(x, F(x), label='Cumulative function', lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#to pruce a set of samples that has the same probability distribution as f(x), generate
#N uniform samples....
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#sampled x values
xsamples = Finv(z);

#how do they compare to the original
hist(xsamples, 50, normed=True, label="Sampled distribution")
plot(x,f(x),label='True distribution', lw=3, color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

Exercise 1. Exponential function - analytic

```python
#integral of lam*exp(-lam*x) = -exp(-lam*x)
#F^(-1)(x)= -ln(lam*x)
x =arange(0.,5.001,0.01)
lam = 1
f = lambda x: lam*exp(-lam*x)
F = lambda x:1 -exp(-lam*x)
invF = lambda x: -log(lam*x -1)

plot(x, f(x), label='True distribution', lw = 3)
plot(x, F(x), label='Cumulative function', lw=3)
#plot(x, invF(x), label='Inverse function', lw=3)
legend(loc=0)
xlim(0,5)
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
invF = lambda x: -log(lam*(1-x))
    
#sampled x values
xsamples = invF(z);

#how do they compare to the original
hist(xsamples, 50, normed=True, label="Sampled distribution")
plot(x,f(x),label='True distribution', lw=3, color='red')

xlim(0,5)
ylim(0,1)

legend(loc=0)
show()
```

```python
x = arange(0.,5.1,0.1)
lam = 1
f = lambda x: lam*exp(-lam*x)
cdf = cumsum(f(x)) / sum(f(x))

cdf
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#sampled x values
xsamples = x[cdf.searchsorted(z)]

#how do they compare to the original
hist(xsamples, 50, normed=True, label="Sampled distribution")
plot(x,f(x),label='True distribution', lw=3, color='red')
xlim(0,5)
legend(loc=0)
show()
```

```python
def distribute1d(x,prob,N):
    F = cumsum(prob)/sum(prob)
    z=numpy.random.random_sample(N);
    T = F.searchsorted(z)
    xsamples = x[T]
    return xsamples

r = distribute1d(x,f(x),10000)
hist(r,50, normed=True)
```
