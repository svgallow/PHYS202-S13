```python
import numpy as np

x = arange(-5,6,1)
A =np.array(cosh(x))
print A

B = np.array(sinh(x))
print B
```

```python
#Plotting sinh and cosh
#subplot(m,n,P) creates an MxN matrix, and places the following plot in the P position 
subplot(1,2,1); plot(x,A)
title('Cosh')
subplot(1,2,2); plot(x,B)
title('Sinh')
 
#ax = gca needed to modify axes
ax = gca()
#deleting boundaries
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
#remove axes of sinh graph
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_position(('data', 0))
ax.spines['right'].set_position(('data', -1))

xlabel("x-axis")
ylabel("y-axis")

```
