```python
import numpy as np
theta = arange(-pi, pi + 0.1, 0.1)
y =  (np.cos(2.*theta)*np.cos(theta/2.))**2.
```

```python
plot(theta, y, linestyle = "-", color = 'r')
plot(theta, y, linestyle = ":", color = 'b', lw = 5)
ylabel("Intensity")
xlabel(r'$\theta$ (rad)')
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
```
