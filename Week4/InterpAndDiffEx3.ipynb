(a) Write a function called finiteDifference(x,y) that takes as arguments arrays of values x and y(x) and returns the derivative of y with respect to x.
-

```python
def finiteDifference(x,y):
    """takes argument arrays of values x and y(x) and returns the derivative of y with respect to x."""
    dydx = zeros(y.shape,float)
    dydx[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

erencing methods to reach a smoother and more accurate derivative.

Test that it works for y = x^5 and plot the percent error from the analytical result.


```python
def fourPtFiniteDiff(x,y):
    """Returns derivative of y w/respect to x using four point method. x and y are arrays"""
    dydx = zeros(y.shape, float)
    dydx[2:-2] = (y[0:-4]-(8*y[1:-3])+(8*y[3:-1])-y[4:])/(12*(x[3]-x[2]))
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

```python
x = arange(0.,10.01, 0.1)
y = x**5
fourPtFiniteDiff(x,y)
```

```python
fourPtDiff = fourPtFiniteDiff(x,y)
percentDiff= 100*abs(fourPtDiff[1:]-dydxAnalytical[1:])/dydxAnalytical[1:]
semilogy(x[1:], percentDiff, label='Percent Error')
legend(loc='upper right')
show()
```
