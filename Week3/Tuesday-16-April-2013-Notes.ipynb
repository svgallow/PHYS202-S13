```python
import numpy as np
t = np.arange(0.0, 1+0.01, 0.01)
s = np.cos(2*2*pi*t)
c = np.sin(4*pi*t)
plot(t, s, linewidth = 2.5, label = "$cos(4\pi\t)$", linestyle = ":")
plot(t, c, color = "red", label = "$sin(4\pi\t)$")
xlabel("time (s)")
ylabel("voltage (mV)")
grid(True)
legend(loc = "lower left")

title("plot", fontsize =30)
savefig('simple_plot.pdf')
```
