```python
i = 1j
#k values here are the squared values of k from the legend
k=0
k1 = 0.25
k2 = 1
k3 = 4

t=np.arange(-pi, pi, 0.0001)
s = (np.tanh(t)**2 + k) / (1 + k)
m = (np.tanh(t)**2 + k1) / (1 + k1)
n = (np.tanh(t)**2 + k2) / (1 + k2)
b = (np.tanh(t)**2 + k3) / (1 + k3)
h= -(1 / np.cosh(t))

plot(t,h, label = "$U($x$)$")
plot(t,s, label = "$\psi^2,k=0$", linestyle = "-")
plot(t,m, label = "$\psi^2,k=0.5$", linestyle = "--")
plot(t,n, label = "$\psi^2,k=1$", linestyle = "-.")
plot(t,b, label = "$\psi^2,k=2$", linestyle = ":")
legend(loc = "lower left")
xlabel("$x$(rad)")
ylabel("U(x),$\psi^2(x)$")
xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
```

```python
data = loadtxt('/Users/Stewart/Desktop/sunspots.txt', float)
x = data[:,0]
y = data[:,1]
   
plot(x[:1000:],y[:1000:])
```

```python
a = np.zeros(1000)
b = np.zeros(1000)
avg = np.zeros(1000)
tot = 0
for i in range(1000):
    a[i] = x[i]
    b[i] = y[i]
    tot += y[i]
    avg[i] = tot / i

    


plot(a, b, lw = 1, label = "Sunspots")
plot(a, avg, lw = 5, label = "Average")
xlabel("Months")
ylabel("# of Sunspots")
title("Sunspots per months since 1749 [First 1000]")
legend()
show()
```
