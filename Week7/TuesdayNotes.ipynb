```python
g = 9.8
l = 0.5
th0 = 20.*pi/180

t = arange(0.,10.01,0.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)
plot(t,theta_app(t))
show()
```

```python
def deriv(theta_current, t):
    g = 9.8
    l = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
from scipy.integrate import odeint
```

```python
t = arange(0.,10.01,0.01)
omega0 = 0.
theta_init = array([th0, omega0])
theta = odeint(deriv, theta_init,t)
```

```python
theta.shape
```

```python
plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

```python
th0 = 0
theta_init = array([th0, omega0])
theta = odeint(deriv, theta_init,t)
plot(t,theta_app(t),'b-', t, theta[:,0], 'r--')

th1 = 45.*pi/180
theta_init1 = array([th1, omega0])
theta1 = odeint(deriv, theta_init1,t)
plot(t,theta_app(t),'b-', t, theta1[:,0], 'o--')

th2 = 90.*pi/180
theta_init2 = array([th2, omega0])
theta2 = odeint(deriv, theta_init2,t)
plot(t,theta_app(t),'b-', t, theta2[:,0], 'g--')

th3 = 135.*pi/180
theta_init3 = array([th3, omega0])
theta3 = odeint(deriv, theta_init3,t)
plot(t,theta_app(t),'b-', t, theta3[:,0], 'b--')

th4 = 180.*pi/180
theta_init4 = array([th4, omega0])
theta4 = odeint(deriv, theta_init4,t)
plot(t,theta_app(t),'b-', t, theta4[:,0], 'r--')

show()
```

When the initial angle is 0, there is no movement.

When the initial angle is 180 degrees, the pendulum remains motionless for almost 10 seconds.
