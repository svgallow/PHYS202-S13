In natural units (where all the constants are 1), the wavefunction of the n-th energy
level of the one-dimensional quantum harmonic oscillator - i.e., a spinless point particle in
a quadratic potential well - is given by

$\psi_n(x)$ = $\frac{1}{\sqrt{2^n n!\sqrt{\pi}}} e^{-x^2/2} H_n(x)$

(a) Write a user-defined function H(n,x) that calculates Hn(x) for given x and any integer n >= 0. Use your function to make a plot that shows the wave functions for n=0, 1, 2, and 3, all on the same graph, in the range x = -4 to x = 4. Hint: There is a function factorial in the math package that calculates the factorial of an integer.

```python
def H(n,x):
    """H(n,x): Returns Hermite polynomial of an integer, n, at any x"""
    if n == 0:
        H = 1
    elif n == 1:
        H = 2*x
    elif n == 2:
        H = (2*x)**2-2
    else:
        H_n = 2*x
        H_last = 1
        H_next = 2*x*H_n-2*1*H_last
        for i in range(2,n):
            H = 2*x*H_next-2*i*H_n
            H_n = H_next
            H_next = H
    return H
print H(3,1)
```

(a) Use the hit or miss Monte Carlo method to estimate the integral of $f(x) = 4\sqrt{1-x^2}$

in the interval 0 $\leq$ x $\leq$ 1 as a function of N, the number of samples.

```python
f = lambda x: 4*(1.-x**2.)**(.5)
x = arange(0.,1.,.01)

xmin = 0.
xmax = 1.
subx = x[logical_and((x > xmin),(x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.
N = 100000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)
newmask = (samples_y < f(samples_x))
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I,'answer with N=100000'
```

(c) Make a log-log plot of the error you calculated in part(b) as a function of N. What is
the approximate functional dependence of the error on N for large N, for example N $\geq$ $10^4$
?

```python
from pylab import *
f = lambda x: 4*(1.-x**2.)**(.5)
N = np.zeros(100000)
z = np.zeros(100000)
n = 1
for i in range(100000):
    samples_x = xmin + (xmax-xmin)*random_sample(i)
    samples_y = ymin + (ymax-ymin)*random_sample(i)
    newmask = (samples_y < f(samples_x))
    I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
    z[i] = ((I-pi)/pi)*100
    N[i] = i
    n += 1
    
print z
plot(N,z)
loglog(N,z)

xlabel("N", fontsize = 20)
ylabel("% Error", fontsize = 20)


#The error is not very dependent on N. The average error around N > 10^4 is approximately the same as N ~ 10^3
```

```python
def Mean(N,xmin,xmax):
    f = lambda x: 4*np.sqrt(1 - x**2)
    xi = linspace(xmin,xmax,N)
    b_a = xmax - xmin
    A = 1./N
    C = b_a * A
    tot = 0.
    for q in xi:
        tot += f(q)
    sumMean = C*tot 
    return sumMean
```

```python
New = arange(1,1000,10)
sample = []

for q in New:
    sample = append(sample, Mean(q,0.,1.))

loglog(New, error(sample), label = 'percent error')
xlabel("N", fontsize = 20)
ylabel("Percent Error", fontsize = 20)
grid(True)
legend(loc = 'lower left')
show()
```
