```python
f = lambda x: x**.5 + sin(x)
```

```python
x = arange(0., 15., .001)
```

```python
myplot = plot(x,f(x), lw = 3)
xlabel(r'$x$', fontsize=20)
ylabel(r'$x^{1/2} + \sin(x)$', fontsize=20 )
xlim(0.,15.)
```

```python
xmin = 1.7
xmax =12.4

#find the maximum value of f within the boundary to set your ymax
subx = x[logical_and((x > xmin), (x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

#Replot the Function
myplot = plot(x,f(x), lw=3)
xlim(0.,15.)

#plot da box
plot([xmin, xmin], [ymin,ymax], color ='k', linestyle = '--') #vertical left
plot([xmax, xmax], [ymin,ymax], color = 'k', linestyle = '--') #vertical right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle ='--') #horizontal top

show()
```

```python
#random_sample?

N = 10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);
    
#Replot the function
myplot = plot(x,f(x), lw=3)
xlim(0.,15.)
    
plot(samples_x[1:200],samples_y[1:200],'.')
    
#plot da box
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') #vert left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vert right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--') #horiz top
    
show()
    
```

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

We can replot the points that lie below the function for proof

```python
#replot the function
myplot = plot(x,f(x), lw=3)
xlim(0., 15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')

#plot da box AGAIN
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') #vert left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vert right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--') #horiz top

show()
```

We can compute our integral from step V: the area of the rectangle times the ration of the number of red and green dots

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

We can evaluate the integral at the two endpoints to ontain the exact solution:

```python
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact=", Iexact
print "I_hitormiss=", I
print "Error on I =", Ierr,"%"
```
